```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data=pd.read_json('ProductDetail.json') 
```

```python
data
```

```python
#Thong ke theo location
sold_value = data.groupby('position').sum()['sold']
posi = [position for position, sold in sold_value.items()]
plt.bar(x=posi, height=sold_value)
plt.xticks(posi, rotation=90, size=6)
plt.xlabel('Position')
plt.ylabel('Sold')
plt.show()
```

```python
#Thong ke theo category
sold_cate = data.groupby('category').sum()['sold']
cate = [category for category, sold in sold_cate.items()]
plt.bar(x=cate, height=sold_cate)
plt.xticks(cate, rotation=90, size=6)
plt.xlabel('Category')
plt.ylabel('Sold')
plt.show()
```

```python
#Bieu do Category
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(sold_cate, labels = cate, autopct='%1.2f%%')
plt.show()
```
